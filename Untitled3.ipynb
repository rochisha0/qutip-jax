```python
from qutip import (about, basis, fmmesolve, fsesolve,
                   plot_expectation_values, sigmax, sigmaz)
```

```python
with qutip.CoreOptions(default_dtype="jax"):
    delta = 0.2 * 2 * jnp.pi
    eps0 = 2 * jnp.pi
    A = 2.5 * 2 * jnp.pi
    omega = 2 * jnp.pi
    H0 = -delta / 2.0 * sigmax() - eps0 / 2.0 * sigmaz()
    H1 = [A / 2.0 * sigmaz(), "sin(w*t)"]
    args = {"w": omega}
    H = [H0, H1]
    psi0 = basis(2, 0)

```

```python
T = 2 * jnp.pi / omega
# simulation time
tlist = jnp.linspace(0, 2.5 * T, 101)
# simulation
result = fsesolve(H, psi0, tlist, T=T, e_ops=[sigmaz()],
                  args=args, options={"method": "diffrax"})

plot_expectation_values([result], ylabels=["<Z>"]);

```

```python
# Noise Spectral Density
gamma = 0.1
def noise_spectrum(omega):
    return gamma * omega / (4 * jnp.pi)
# Coupling Operator
with qutip.CoreOptions(default_dtype="jax"):
    c_ops = [sigmax()]

# Solve using Fmmesolve

fme_result = fmmesolve(
    H,
    psi0,
    tlist,
    c_ops=c_ops,
    spectra_cb=[noise_spectrum],
    e_ops=[sigmaz()],
    T=T,
    args=args,
    options={"method": "diffrax"}
)
```

```python
N = 15
w0 = 1.0 * 2 * jnp.pi
A = 0.1 * 2 * jnp.pi
times = jnp.linspace(0, 10, 201)
gamma = 0.25
ntraj = 50
with qutip.CoreOptions(default_dtype="jax"):
    a = destroy(N)
    x = a + a.dag()
    y = -1.0j * (a - a.dag())
    H = w0 * a.dag() * a + A * (a + a.dag())
    rho0 = coherent(N, np.sqrt(5.0), method="analytic")
    c_ops = [np.sqrt(gamma) * a]
    e_ops = [a.dag() * a, x, y]
```

```python
result_ref = mesolve(H, rho0, times, c_ops, e_ops, options={"method": "diffrax"})
```

```python
plot_expectation_values(result_ref);
```

```python
options = {"store_measurement": True, "map": "parallel", "method": "diffrax"}

result = smesolve(
    H,
    rho0,
    times,
    sc_ops=c_ops,
    heterodyne=True,
    e_ops=e_ops,
    ntraj=ntraj,
    options=options
)
```

```python
import qutip_jax
import qutip as qt
from qutip.solver.sesolve import sesolve, SESolver
import jax
@jax.jit
def fp(t, w):
    return jax.numpy.exp(1j * t * w)

@jax.jit
def fm(t, w):
    return jax.numpy.exp(-1j * t * w)

H = (
    qt.num(2)
    + qt.destroy(2) * qt.coefficient(fp, args={"w": 3.1415})
    + qt.create(2) * qt.coefficient(fm, args={"w": 3.1415})
).to("jax")

ket = qt.basis(2, 1)

solver = SESolver(H, options={"method": "diffrax"})

def f(w, solver):
    result = solver.run(ket, [0, 1], e_ops=qt.num(2).to("jax"), args={"w":w})
    return result.e_data[0][1].real

jax.grad(f)(0.5, solver)
```
